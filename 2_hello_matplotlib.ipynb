### 基础图像绘制

```python
# 0.数据准备
x = range(60)
y_sz = [random.uniform(25, 35) for i in x]

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.plot(x, y_sz)

# 3.图像显示
plt.show()
```

###  完善图像

```python
# 0.数据准备
x = range(60)
y_sz = [random.uniform(25, 35) for i in x]

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.plot(x, y_sz)

# 2.1 添加x,y轴刻度
y_ticks = range(40)
x_ticks = ["11点{}分".format(i) for i in x]

# 2.2 添加网格显示
plt.grid(linestyle="--", alpha=0.9)

# 2.3 添加x,y轴刻度+标题
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("深圳一小时温度变化图", fontsize=20)

plt.yticks(y_ticks[::5])
plt.xticks(x[::5], x_ticks[::5])

# 2.4 图像保存
plt.savefig("温度变化图.png")

# 3.图像显示
plt.show()

# # 2.4 图像保存  # 如果放到了show的后面,会进行资源的释放!
# plt.savefig("温度变化图.png")
```

### 一个坐标系下绘制多个图像

```python
# 0.数据准备
x = range(60)
y_sz = [random.uniform(25, 35) for i in x]
y_bj = [random.uniform(15, 20) for i in x]

# 1. 创建画布
plt.figure(figsize=(20, 8), dpi=100)

# 2.绘制图像
plt.plot(x, y_sz, label="深圳")
plt.plot(x, y_bj, linestyle="-.", color="r", label="北京")

# 2.1 添加x,y轴刻度
y_ticks = range(40)
x_ticks = ["11点{}分".format(i) for i in x]

# 2.2 添加网格显示
plt.grid(linestyle="--", alpha=0.9)

# 2.3 添加x,y轴刻度+标题
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("一小时温度变化图", fontsize=20)

plt.yticks(y_ticks[::5])
plt.xticks(x[::5], x_ticks[::5])

# 2.4 图像保存
plt.savefig("温度变化图.png")

# 2.5 添加图例
plt.legend(loc=0)

# 3.图像显示
plt.show()
```
